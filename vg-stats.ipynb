# Lab 06: Aggregate Analysis - Video Game Stats - Jay Adams

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/vgsales.csv')
```

```python
df.head()
```

# Which company is the most common video game publisher?

```python
common = pd.DataFrame(df['Publisher'].groupby(df.Publisher).count())
common.sort_values('Publisher', ascending=False).head()
```

# What’s the most common platform?

```python
common = pd.DataFrame(df['Platform'].groupby(df.Platform).count())
common.sort_values('Platform', ascending=False).head()
```

# What about the most common genre?

```python
common = pd.DataFrame(df['Genre'].groupby(df.Genre).count())
common.sort_values('Genre', ascending=False).head()
```

# What are the top 20 highest grossing games?

```python
best_sales = df[['Name', 'Global_Sales']].sort_values('Global_Sales', ascending=False).head(20)
```

## For the top-selling game of all time, how many standard deviations above/below the mean are its sales for North America?

```python
df_mean = np.mean(df['NA_Sales'])
df_std = pd.DataFrame(df[df.NA_Sales.values >= df_mean])
df_std.NA_Sales.std()
```

## The Nintendo Wii seems to have outdone itself with games. How does its average number of sales compare with all of the other platforms?

```python
df_wii = df[['Platform', 'Global_Sales']].groupby('Platform')
df_wii = pd.DataFrame(df_wii['Global_Sales'].mean()).sort_values('Global_Sales', ascending=False)
df_wii
```
